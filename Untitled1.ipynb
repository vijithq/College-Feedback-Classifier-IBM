```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
import pandas as pd

# Load FLAN-T5 model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-small")
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-small")

# Create pipeline
pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer)
# Load the uploaded file (make sure to match the filename)
df = pd.read_csv("college_feedback_50.csv")  # Replace with actual name if different

# Define the classification function using the model
def classify_feedback(text):
    prompt = f"Classify the following student feedback into one of the categories: Academics, Facilities, or Administration.\nFeedback: {text}"
    result = pipe(prompt, max_length=50, clean_up_tokenization_spaces=True)[0]['generated_text']
    return result.strip()
# Apply the classification model
df['feedback_category'] = df['feedback_text'].apply(classify_feedback)
# Show first 10 classified feedbacks
df[['student_name', 'feedback_text', 'feedback_category']].head(10)
```

```python
import matplotlib.pyplot as plt

# Count number of feedbacks in each category
category_counts = df['feedback_category'].value_counts()

# Plot the bar chart
plt.figure(figsize=(8, 5))
plt.bar(category_counts.index, category_counts.values, color='skyblue')
plt.title("Number of Feedbacks per Category")
plt.xlabel("Feedback Category")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
